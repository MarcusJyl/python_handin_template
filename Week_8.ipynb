```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv('./data/cars.csv')
con_str = "mysql+mysqlconnector://root:root@db/db"
engine = create_engine(con_str)
#connection = engine.raw_connection()
df.to_sql('cars_db',con=engine, if_exists='append', index = False)



```

# Endpoints


```python

#!flask/bin/python
from flask import Flask, jsonify, abort, request
import mysql.connector as mysql

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello, World!'

@app.route('/api/car/<int:id>', methods=['GET'])
def get_by_id(id):
    ## connecting to the database using 'connect()' method
    cnx = mysql.connect(host = "db", user = "root", passwd = "root", db = "db")
    # cursor = cnx.cursor()

    curA = cnx.cursor()
    

    
    query = ("SELECT * FROM cars_db WHERE Id = %s;")


    
    curA.execute(query, (id,))
    result = curA.fetchall()
    return str(result)

@app.route('/api/car', methods=['GET'])
def get_all():
    ## connecting to the database using 'connect()' method
    cnx = mysql.connect(host = "db", user = "root", passwd = "root", db = "db")
    # cursor = cnx.cursor()

    curA = cnx.cursor()
    

    
    query = ("SELECT * FROM cars_db;")


    
    curA.execute(query)
    result = curA.fetchall()
    return jsonify(result)
    
@app.route('/api/car/brand/<brand>', methods=['GET'])
def get_by_brand(brand):
    ## https://code-maven.com/flask-routes
    ## connecting to the database using 'connect()' method
    cnx = mysql.connect(host = "db", user = "root", passwd = "root", db = "db")
    # cursor = cnx.cursor()

    curA = cnx.cursor()
    

    
    query = ("SELECT * FROM cars_db WHERE make = %s;")


    
    curA.execute(query, (brand,))
    result = curA.fetchall()
    return jsonify(result)

@app.route('/api/car', methods=['POST'])
def add_car():
    ## connecting to the database using 'connect()' method
    cnx = mysql.connect(host = "db", user = "root", passwd = "root", db = "db")
    # cursor = cnx.cursor()

    curA = cnx.cursor(prepared=True)
    

    
    query = ("INSERT INTO cars_db (Make, Model, Type, Origin, DriveTrain, MSRP, Invoice, EngineSize, Cylinders, Horsepower, MPG_City, MPG_Highway, Weight, Wheelbase, Length) VALUES (%s, %s, %s ,%s, %s, %s ,%s, %s, %s ,%s, %s, %s ,%s, %s, %s);")
    val = (request.json['Make'], request.json['Model'], request.json['Type'], request.json['Origin'], request.json['DriveTrain'], request.json['MSRP'], request.json['Invoice'],request.json['EngineSize'],request.json['Cylinders'],request.json['Horsepower'], request.json['MPG_City'], request.json['MPG_Highway'], request.json['Weight'], request.json['Wheelbase'], request.json['Length'])
    
    
    curA.execute(query, val)
    cnx.commit()
    result = str(curA.lastrowid) + " record inserted."
    curA.close()
    cnx.close()
    
    return jsonify(result)

@app.route('/api/car/<int:id>', methods=['DELETE'])
def delete_car(id):
    ## connecting to the database using 'connect()' method
    cnx = mysql.connect(host = "db", user = "root", passwd = "root", db = "db")
    # cursor = cnx.cursor()

    curA = cnx.cursor(prepared=True)
    

    
    query = ("DELETE FROM cars_db WHERE Id = %s;")


    
    curA.execute(query, (id,))
    cnx.commit()
    result = str(id) + " deleted."
    curA.close()
    cnx.close()
    
    return jsonify(result)
```
